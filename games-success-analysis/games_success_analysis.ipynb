# Games Success Analysis

## Project Description

The "Strimchik" online store sells computer games all over the world. Historical game sales data, user and expert ratings, genres and platforms (such as Xbox or PlayStation) are available from public sources. You need to identify the patterns that determine the success of the game. This will allow you to bid on a potentially popular product and plan advertising campaigns.

# Data Description

- `Name` — game name

- `Platform` — platform

- `Year_of_Release` — release year

- `Genre` - game genre

- `NA_sales` - sales in North America (millions of copies sold)

- `EU_sales` - sales in Europe (millions of copies sold)

- `JP_sales` - sales in Japan (millions of copies sold)

- `Other_sales` - sales in other countries (millions of copies sold)

- `Critic_Score` - score of critics (maximum 100)

- `User_Score` - user score (maximum 10)

- `Rating` — rating from the ESRB (Entertainment Software Rating Board). This association determines the rating of computer games and assigns them an appropriate age category.

## Data Preprocessing

### Column names

```python
# lowercase

df.columns = [x.lower() for x in df.columns]
df.columns
```

### Change data formats

```python
df['year_of_release'] = df['year_of_release'].fillna(0)
df['year_of_release'] = df['year_of_release'].astype('int')

df = df.replace('tbd', np.nan)
df['user_score'] = df['user_score'].astype('float')

df.info()
```

```python
df['year_of_release'].fillna(0).astype('int')
```

### Missing values

#### `name` &  `genre`

```python
df = df[df['name'].notnull()]

df.info()
```

#### `year_of_release`

```python
df['year_of_release'] = df['year_of_release'].replace(0, np.nan)
df['year_of_release'] = df.groupby(['name', 'platform']).transform(lambda x: x.fillna(x.mean()))['year_of_release']
df['year_of_release'].value_counts().sort_values(ascending=False)
```

```python
df[df['year_of_release'].isnull()].head(10)
```

Not much - delete the rows

```python
df = df[df['year_of_release'].notnull()]
df['year_of_release'] = df['year_of_release'].astype('int')
df.info()
```

## Exploratory Data Analysis

### Games through years

```python
fig, ax = plt.subplots(figsize = (13,10))
ax = sns.distplot(df['year_of_release'], bins=len(df['year_of_release'].unique()), color='midnightblue', \
                  label='Распределение объема выпускаемых игр по годам')

plt.suptitle('Распределение числа игр по годам', fontsize=18)
plt.xticks(range(df['year_of_release'].sort_values().unique()[0], \
                 df['year_of_release'].sort_values().unique()[-1]), rotation='vertical')
```

In general, you can see that a large number of games were released from 2002 to 2015, with a peak occurring in the period 2004-2011.
This tells us that the data that is before 2002 can simply not be considered, but this is not enough to select the current analysis period - we need to look at the platforms.

It can also be noted that the gaming industry, based on the data provided to us, is undergoing a rather serious decline over the past 5 years - we can observe a sharp jump down in game sales.

### Sales changes by platform

```python
# Let's take a new period starting from 2002, which we decided to consider based on our previous conclusions regarding
# number of released games until 2016

df2002 = df.query("year_of_release >= 2002").copy()

years_platforms = df2002.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')

plt.rcParams['figure.figsize'] = [20,10]
years_platforms_cols = years_platforms.columns.tolist()
for col in years_platforms:
    plt.plot(years_platforms[col], label=col, linewidth=2)

plt.title('Суммарные продажи по платформам с 2002 г.', fontsize=18)
plt.xlabel('Years', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.legend()

plt.show()
```

Little is clear from the graph - we took too many platforms in the sample for the graph. Moreover, they differ much in  total sales. Not the best visualization solution to come to clear conclusion.

However, in general, it can be seen that each of the platforms has the same distribution of games sold. The launch date of each of them is accompanied by a peak in sales for about 1 year and then a smooth decline is spotted until the platform becomes completely obsolete.

Let's select the platforms with the top highest total sales in order to see in more detail what happens with the distribution of sales of the platforms at their picks

```python
platfom_total_sales = df2002.pivot_table(index='platform', values='total_sales', aggfunc='sum')
platfom_total_sales.columns = ['total_sales']

platfom_total_sales.sort_values('total_sales', ascending=False)
```

Thus, the platforms with the highest total sales (since 2002) are: **X360, PS3, Wii, PS2, DS**

Let us study the exact time period for new platforms to appear and old ones to disappear.

```python
fig, axes = plt.subplots(2,3)

axes[0,0].plot(years_platforms['X360'], label='X360', linewidth=2, color='red')
axes[0,1].plot(years_platforms['PS3'], label='PS3', linewidth=2, color='orange')
axes[0,2].plot(years_platforms['Wii'], label='Wii', linewidth=2, color='yellow')
axes[1,0].plot(years_platforms['PS2'], label='PS2', linewidth=2, color='green')
axes[1,1].plot(years_platforms['DS'], label='DS', linewidth=2, color='blue')

axes[0,0].set_title('X360', fontsize=14)
axes[0,1].set_title('PS3', fontsize=14)
axes[0,2].set_title('Wii', fontsize=14)
axes[1,0].set_title('PS2', fontsize=14)
axes[1,1].set_title('DS', fontsize=14)

axes[0,0].set_xlabel("years", fontsize=12)
axes[0,0].set_ylabel("total_sales", fontsize=12)
axes[0,1].set_xlabel("years", fontsize=12)
axes[0,1].set_ylabel("total_sales", fontsize=12)
axes[0,2].set_xlabel("years", fontsize=12)
axes[0,2].set_ylabel("total_sales", fontsize=12)
axes[1,0].set_xlabel("years", fontsize=12)
axes[1,0].set_ylabel("total_sales", fontsize=12)
axes[1,1].set_xlabel("years", fontsize=12)
axes[1,1].set_ylabel("total_sales", fontsize=12)

fig.suptitle('Продажи на топовым платформ по годам с 2002 г. (раздельно)', fontsize=20)
plt.subplots_adjust(hspace=0.4)
plt.show()
```

```python
top_platforms = ['X360', 'PS3', 'Wii', 'PS2', 'DS']

for platf in top_platforms:
    plt.plot(years_platforms[platf], label=platf, linewidth=3)

plt.title('Продажи топовых платформ по годам с 2002 г. (совместно)', fontsize=18)    
plt.xlabel('Years', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.legend()
```

Analyzing the platforms with the largest total sales since 2002, we can say that new platforms appear at a frequency of approximately **1 time per year** since 2004. Also, this characteristic is accompanied by a large increase in the volume of released games, which we could see on the chart on the distribution of game releases above. So it can be argued that these events are conjugated. That is, at the peak of the gaming industry, which we identified earlier, the speed of release of new platforms is extremely high, as well as the volume of new games being created.

To determine the approximate lifespan (the amount of time for which the platform disappears), I propose to consider exactly the part of each of the distributions where the platform is at the peak of its profitability + some high part on the left and right. This would be logical, since, as we have already stated, a new platform comes out once a year and the old ones are already beginning to lose their relevance and popularity. Consequently, you can say that the platforms disappear in about **3 years**.
(If we consider in general how long the platform lives before its complete disappearance, then this period is much longer - about 10 years).

One important trend should also be taken into account. Just as we looked at the volume distribution of games earlier and said that games are being produced much less over time (since 2011), looking at the distribution of sales by platform, we can say that new platforms have taken much longer to be produced. Since 2012, this **term has increased to 2 years**. And since 2014, according to our data, there are no newly produced platforms at all.

## Hypothesis Testing

### Hypothesis: The mean user ratings of the Xbox One and PC platforms are the same

```python
# create datasets to compare 2 platforms

df_rating = df_actual_upd[df_actual_upd['user_score'].notnull()].copy()

xone = df_rating[df_rating['platform'] == 'XOne'].copy()
pc = df_rating[df_rating['platform'] == 'PC'].copy()
```

Let's start with the **Levene's Test** to see if the hypothesis that the two analyzed samples are taken from populations with equal variances is correct.
We do this in order to correctly form the premise for our subsequent test.

```python
# Levene's test

alpha = 0.05

levene_platf = st.levene(xone['user_score'], pc['user_score']) 

print('\033[1m' + 'Сравнение дисперсий совокупностей 2-х платформ' + '\033[0m')
print('--------------------------')
print('\033[1m' + 'p-value:' + '\033[0m', levene_platf.pvalue)
print()

if (levene_platf.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, что дисперсии совокупностей 2-х тарифов равны")
else:
    print("Не получилось отвергнуть нулевую гипотезу о равенстве совокупностей 2-х тарифов")
```

---

$ x = Xbox One $

$ p = PC $

$$H_0: μ_x = μ_p  $$ 
$$H_A: μ_x \neq μ_p  $$ 

$$\alpha = 0.05 $$

*Assumption: populations have **various** variances*

We will use the **two-sided t-test** for the mean of two independent samples, since we are interested in the deviation both in one direction and in the other.

---


**Assumption**: Population variances are different.

```python
# t-test:

rate_platf = st.ttest_ind(xone['user_score'], pc['user_score'], equal_var=False) 

print('\033[1m' + 'Сравнение средних пользовательских оценок совокупностей 2-х платформ' + '\033[0m')
print('--------------------------')
print('\033[1m' + 'p-value:' + '\033[0m', rate_platf.pvalue)

if (rate_platf.pvalue < alpha):
    print("Отвергаем нулевую гипотезу о равенстве средних пользовательских оценок совокупностей 2-х платформ")
else:
    print("Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских оценок совокупностей 2-х платформ")
```

#### Conclusion

We failed unable to reject the null hypothesis that the average user ratings of the **Xbox One** and **PC** platforms are the same. Moreover, we would also not be able to reject this hypothesis at a significance level = 1%

### Hypothesis: The average user ratings of the Action and Sports genres are different

```python
# Let's create datasets to compare 2 genres

action = df_rating[df_rating['genre'] == 'Action'].copy()
sports = df_rating[df_rating['genre'] == 'Sports'].copy()
```

```python
# Levene's test

alpha = 0.05

levene_genre = st.levene(action['user_score'], sports['user_score']) 

print('\033[1m' + 'Сравнение дисперсий совокупностей 2-х жанров' + '\033[0m')
print('--------------------------')
print('\033[1m' + 'p-value:' + '\033[0m', levene_genre.pvalue)
print()

if (levene_genre.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, что дисперсии совокупностей 2-х жанров равны")
else:
    print("Не получилось отвергнуть нулевую гипотезу о равенстве совокупностей 2-х жанров")
```

---

We will use the **two-sided t-test** for the mean of two independent samples, since we are interested in the deviation both in one direction and in the other.

---


**Assumption**: Population variances are different.

$ a = Action $

$ s = Sports $

$$H_0: μ_a = μ_s  $$ 
$$H_A: μ_a \neq μ_s  $$ 

$$\alpha = 0.05 $$

*Assumption: populations have **various** variances*

```python
# t-test:

rate_genre = st.ttest_ind(action['user_score'], sports['user_score'], equal_var=False) 

print('\033[1m' + 'Сравнение средних пользовательских оценок совокупностей 2-х жанров' + '\033[0m')
print('--------------------------')
print('\033[1m' + 'p-value:' + '\033[0m', rate_genre.pvalue)

if (rate_genre.pvalue < alpha):
    print("Отвергаем нулевую гипотезу о равенстве средних пользовательских оценок совокупностей 2-х жанров")
else:
    print("Не получилось отвергнуть нулевую гипотезу о равенстве средних пользовательских оценок совокупностей 2-х жанров")
```

#### Conclusion

We reject the null hypothesis that the average user ratings of the **Action** and **Sports** genres are the same. That is, we have enough statistical evidence that the average user ratings of the selected genres differ by significance level = 5%
